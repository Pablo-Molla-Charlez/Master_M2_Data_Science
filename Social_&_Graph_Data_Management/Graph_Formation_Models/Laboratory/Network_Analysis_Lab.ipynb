## 1.4 Compute the diameter of the graph and the average distance in the graph.

### Diameter of the Graph

The $\textcolor{red}{\text{diameter}}$ $d_{\text{max}}$ of a graph is the maximum distance between any pair of nodes in the graph:

$$
d_{\text{max}} = \max_{i,j \in V} d_{ij}
$$

This metric represents the $\textcolor{red}{\text{longest shortest path in the graph}}$.

```python
source = random.choice(list(G.nodes))
shortest_paths = nx.single_source_shortest_path(G, source)

# All the longest shortest paths of the graph G and their lengths
max_path_length = max(len(path) for path in shortest_paths.values())
longest_paths = [path for path in shortest_paths.values() if len(path) == max_path_length]

# One of the longest paths (if there are many) for visualization
l_s_path = longest_paths[0]
l_s_path_edges = list(zip(l_s_path, l_s_path[1:]))
pos = nx.spring_layout(G)
nx.draw(G, pos=pos, with_labels=True)
nx.draw_networkx_edges(G, pos=pos, edgelist=l_s_path_edges, edge_color='r')
plt.show()

print(f"Diameter of the graph: {nx.diameter(G)}")
print(f"Total number of longest shortest paths: {len(longest_paths)}")
```

### Average Distance of the Graph

The $\textcolor{orange}{\text{average distance}}$ $\langle d \rangle$ in a graph is the mean of all pairwise distances between nodes:

$$
\langle d \rangle = \frac{1}{N(N - 1)} \sum_{\substack{i, j \in V \\ i \neq j}} d_{ij}
$$

where $N = |V|$ is the total number of nodes and the summation considers all unique pairs of distinct nodes.

## 1.5 Plot the distribution of the PageRank values in the graph.

### PageRank

The $\textcolor{magenta}{\text{PageRank algorithm}}$ is a method for $\textcolor{magenta}{\text{ranking the importance of nodes in a directed graph}}$ based on their incoming links. It operates by assigning each node an initial score and then iteratively updating these scores by distributing a node's rank through its outbound links. A damping factor is included to account for the probability that a random walker might jump to any node in the graph at any step. Over multiple iterations, the algorithm converges to stable PageRank values, where $\textcolor{magenta}{\text{higher scores indicate more influential or important nodes}}$. Nodes that are connected to well-connected nodes also get a higher score, even if they themselves don't have many connections. Originally $\textcolor{cyan}{\text{developed by Google}}$ to rank web pages, PageRank effectively identifies key nodes within complex networks by considering both the quantity and quality of their connections.

```python
pagerank_dist = [pagerank for node, pagerank in nx.pagerank(G).items()]
deg_seq = [G.degree(x) for x in nx.nodes(G)]

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].hist(pagerank_dist, bins=10)
axes[0].set_title('PageRank Distribution Histogram')
axes[0].set_xlabel('PageRank')
axes[0].set_ylabel('Frequency')

axes[1].hist(deg_seq, bins=10)
axes[1].set_title('Degree Distribution Histogram')
axes[1].set_xlabel('Degree')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

Although $\textcolor{magenta}{\text{PageRank scores do not scale directly with node degrees}}$, their distributions exhibit similarities. Both $\textcolor{magenta}{\text{distributions are right-skewed}}$, indicating that $\textcolor{magenta}{\text{most nodes have low PageRank scores and low degrees}}$. Additionally, they follow a $\textcolor{orange}{\text{power-law distribution}}$, resulting in a $\textcolor{orange}{\text{long tail}}$. This pattern signifies the $\textcolor{magenta}{\text{existence of hubs}}$ (specifically, nodes $\textcolor{magenta}{\text{0}}$ and $\textcolor{magenta}{\text{33}}$) which possess the highest PageRank scores (0.09‚Äì0.10) and degree values (16). To determine the probability $p$ in a random network, we can use the following equation and derive from it:
$$ \langle k \rangle = p \times (N - 1) $$
From this relationship, we can solve for $p$ as follows:
$$ \langle k \rangle = p \times (N - 1) \ \longleftrightarrow \ p = \frac{\langle k \rangle}{N - 1} $$

```python
number_of_edges = len(G.edges())
number_of_nodes = len(G.nodes())
average_degree = 2 * number_of_edges / number_of_nodes

p = average_degree / (number_of_nodes - 1)

print(f"Number of nodes in the graph: {number_of_nodes}")
print(f"Number of edges in the graph: {number_of_edges}")
print(f"Average degree of the graph: {average_degree}")
print(f"Calculated p: {p}")
```

# 3. Counting Triangles in Graphs

## Triangles in Graphs

A $\textcolor{green}{\text{triangle}}$ in a graph is a $\textcolor{green}{\text{set of three}}$ $\textcolor{green}{\text{nodes where}}$ $\textcolor{green}{\text{each node is}}$ $\textcolor{green}{\text{directly connected}}$ $\textcolor{green}{\text{to the other two}}$, forming a closed loop of three edges. The $\textcolor{red}{\text{task}}$ is to $\textcolor{red}{\text{count the number}}$ $\textcolor{red}{\text{of triangles}}$ $\textcolor{red}{\text{in the graph}}$, without using the NetworkX implementation

## 3.1 Compute the number of triangles in which each node is involved. The output should be a dictionary containing the node as a key and the number of triangles as a value.

```python
# Initializing the dictionary to hold nodes and their neighbors
nodes_and_neighbors = {}

# Reading the graph data from the file and building the adjacency list
with open(file='/Users/pablomollacharlez/Desktop/M2_DS/SGDM/karate.txt', mode='r') as file:
    for line in file.readlines():
        node, neighbor = line.strip().split()
        if node not in nodes_and_neighbors:
            nodes_and_neighbors[node] = [neighbor]
        else:
            nodes_and_neighbors[node].append(neighbor)

# Printing neighbors for each node (currently prints only the first node due to 'break')
for node, neighbors in nodes_and_neighbors.items():
    print(f"Node {node} has following neighbors: {neighbors}")
    break

# Starting timing for the triangle calculation
start_time_1 = time.time()

# Initializing the dictionary to hold triangle counts for each node
triangles = {}

# Iterating through each node to count the number of triangles it is part of
for node, neighbors in nodes_and_neighbors.items():
    counter = 0
    # If the node has only one neighbor, it cannot form a triangle
    if len(neighbors) == 1:
        triangles[node] = counter
        continue
    
    # Check each pair of neighbors to see if they are connected
    for i in range(len(neighbors)):
        for j in range(len(neighbors)):
            if i < j:
                if neighbors[j] in nodes_and_neighbors[neighbors[i]]:
                    counter += 1
                else:
                    continue
    triangles[node] = counter

# Ending timing for the triangle calculation
end_time_1 = time.time()
manual_time = end_time_1 - start_time_1

# Printing the time taken to calculate triangles
print(f"Time taken to calculate triangles manually: {manual_time} seconds")

# Loop to print the number of triangles each node is part of
print("\nNumber of triangles each node is part of:")
for node, count in triangles.items():
    print(f"Node {node}: {count} triangle(s)")
```

## 3.2 Comparison of the output from 3.1 with that of the networkx.triangles method.

```python
# Starting timing for the triangle calculation with nx.triangles
start_time_2 = time.time()

# Calculating triangles with nx.triangles
nx_triangles = nx.triangles(G)

# Ending timing for the triangle calculation with nx.triangles
end_time_2 = time.time()

# Computing difference for actual time execution
nx_time = end_time_2 - start_time_2


nx_triangles = dict(sorted(nx_triangles.items(), key=lambda item: int(item[0])))

print(f'Manual triangles calculation time: {manual_time} seconds')
print(f'NetworkX triangles calculation time: {nx_time} seconds')

print()
print("-" * 52)
print(f"{'Node':<5} {'Triangles (Manual)':<25} {'Triangles (NetworkX)':<25}")
print("-" * 52)

for node in sorted(triangles, key=int):
    manual_count = triangles[node]
    nx_count = nx_triangles[str(node)]
    print(f"{node:<5} {manual_count:<25} {nx_count:<25}")
```

## 3.4 Compute the expected number of triangles in a random graph having the same average degree. How does it compare to the one found in reality?

To calculate the $\textcolor{green}{\text{expected number of triangles}}$ in a $\textcolor{magenta}{\text{Random Graph}}$ first, we need to know how many sets of 3 nodes can be chosen from a graph with $n$ nodes. Using the $\textcolor{cyan}{\text{combinatorial formula}}$ we can find the number of ways to choose any 3 nodes from $n$ nodes:

$$C_n^k = \frac{n!}{k!(n-k)!}$$

In our case:

$$C_n^3 = \frac{n!}{3!(n-3)!} = \frac{n(n-1)(n-2)}{6}$$ 

For $\textcolor{green}{\text{three nodes to}}$ $\textcolor{green}{\text{form a triangle}}$, $\textcolor{green}{\text{all three edges}}$ connecting these nodes $\textcolor{green}{\text{must exist}}$. The $\textcolor{green}{\text{probability that}}$ $\textcolor{green}{\text{all three}}$ $\textcolor{green}{\text{edges exist}}$ simultaneously is:
$$p \times p \times p = p^3$$
Since the presence of each edge is independent, we can deduce that the final formula for calculating the $\textcolor{green}{\text{expected number}}$ $\textcolor{green}{\text{of triangles}}$ in a $\textcolor{magenta}{\text{Random Graph}}$ is the following:
$$\mathbb{E}(T) = \frac{n(n-1)(n-2)}{6} \cdot p^3$$

```python
expected_triangles = p ** 3 * number_of_nodes * (number_of_nodes - 1) * (number_of_nodes - 2) // 6
print(f"Expected number of triangles in the graph: {int(expected_triangles)}")
print(f"Actual number of triangles in the graph: {total_triangles}")
```

$\fbox{\textbf{Remark:}}$

From the results, we can see that the $\textcolor{red}{\text{actual triangle}}$ $\textcolor{red}{\text{count is}}$ $\textcolor{red}{\text{almost three}}$ $\textcolor{red}{\text{times higher}}$ $\textcolor{red}{\text{than the}}$ $\textcolor{red}{\text{expected count}}$. This means that the $\textcolor{green}{\text{network from}}$ $\textcolor{green}{\text{the graph}}$ $\textcolor{green}{\text{is organized}}$ $\textcolor{green}{\text{into clusters}}$, where nodes are densely connected, $\textcolor{green}{\text{forming multiple}}$ $\textcolor{green}{\text{triangles within}}$ $\textcolor{green}{\text{each group}}$. On the contrary, the expected count suggests that, in a $\textcolor{magenta}{\text{Random Network}}$, $\textcolor{magenta}{\text{nodes would}}$ $\textcolor{magenta}{\text{be much less}}$ $\textcolor{magenta}{\text{organized into}}$ $\textcolor{magenta}{\text{such connected groups}}$.

### 4.1 Community Detection Method: Girvan-Newman Algorithm

The $\textcolor{magenta}{\text{Girvan-Newman algorithm}}$ is a $\textcolor{magenta}{\text{divisive community}}$ $\textcolor{magenta}{\text{detection method}}$ used $\textcolor{magenta}{\text{to identify}}$ $\textcolor{magenta}{\text{distinct communities}}$ $\textcolor{magenta}{\text{within a network}}$. 

* It operates by $\textcolor{magenta}{\text{iteratively removing}}$ $\textcolor{magenta}{\text{edges with the highest}}$ $\textcolor{magenta}{\text{betweenness centrality}}$. By systematically eliminating these critical edges, the algorithm progressively splits the network into smaller, well-defined communities. 

* At each step of this process, the algorithm calculates the modularity of the current partitioning, which quantifies the strength of the community structure. The $\textcolor{red}{\text{optimal community}}$ $\textcolor{red}{\text{division is}}$ $\textcolor{red}{\text{determined by}}$ $\textcolor{red}{\text{selecting the}}$ $\textcolor{red}{\text{partition that}}$ $\textcolor{red}{\text{achieves the}}$ $\textcolor{red}{\text{maximum modularity}}$. 

* Additionally, once the optimal communities are identified, they can be visualized by coloring each community with a unique color, highlighting the modular organization of the original graph.

### 4.2 Definitions: $\textcolor{orange}{\text{Betweenness Centrality}}$, $\textcolor{green}{\text{Expected Probability of}}$ $\textcolor{cyan}{\text{Linked Nodes}}$, $\textcolor{cyan}{\text{Community Modularity}}$ $\textcolor{cyan}{\text{and Graph Modularity}}$

#### 1. $\textcolor{orange}{\text{Betweenness Centrality}}$

The $\textcolor{orange}{\text{betweenness centrality}}$ $C_B(v)$ of a node $v$ is defined as:
$$
C_B(v) = \sum_{s \neq v \neq t} \frac{\sigma_{st}(v)}{\sigma_{st}}
$$
where $\sigma_{st}$ is the total number of shortest paths from node s to node t,
and $\sigma_{st}(v)$ is the number of those shortest paths that pass through node $v$. Betweenness centrality $\textcolor{orange}{\text{measures how}}$ $\textcolor{orange}{\text{often a node}}$ $\textcolor{orange}{\text{serves as}}$ $\textcolor{orange}{\text{a bridge}}$ $\textcolor{orange}{\text{along the}}$ $\textcolor{orange}{\text{shortest path}}$ $\textcolor{orange}{\text{between other nodes}}$.

#### 2. $\textcolor{green}{\text{Expected Probability}}$ $\textcolor{green}{\text{of Linked Nodes}}$

The $\textcolor{green}{\text{expected probability}}$ $p_{ij}$ for two nodes to be connected, given their degrees $k_i$ and $k_j$, is:
$$p_{ij} = \frac{k_i k_j}{2L}$$

#### 3. $\textcolor{cyan}{\text{Community Modularity and Graph Modularity}}$

The $\textcolor{cyan}{\text{Community Modularity}}$ ($M$) is a metric used $\textcolor{cyan}{\text{to evaluate}}$ $\textcolor{cyan}{\text{the quality}}$ $\textcolor{cyan}{\text{of a}}$ $\textcolor{cyan}{\text{particular division}}$ $\textcolor{cyan}{\text{of a}}$ $\textcolor{cyan}{\text{network into}}$ $\textcolor{cyan}{\text{communities}}$ (or clusters). The goal is to maximize the number of edges within communities while minimizing the edges between different communities. It is computed as:

$$ M_C = \frac{1}{2L} \sum_{(i,j) \in C} (A_{ij} - p_{ij})$$

Where the parameters are:
* $L$ is the total number of edges in the graph.
* $A_{ij}$ if there is an edge between nodes $i$ and $j$; otherwise, it's 0.
* $p_{ij}$ is the expected number of edges between nodes $i$ and $j$ if edges were placed randomly.

To calculate the $\textcolor{blue}{\text{Graph Modularity}}$ (for the entire graph), we sum the modularities of all communities:
$$M = \sum_{C} M_C$$


* $\fbox{\text{Step 1:}}$ We will use the betweenness centrality of each edge; use the NetworkX implementation as exemplified in the notebook. 

* $\fbox{\text{Step 2:}}$  For the clustering, it will be removed the edge having the highest centrality, and recomputed the betweenness centrality. 

* $\fbox{\text{Step 3:}}$ In terms of communities, the communities correspond to the connected components of the graph; these will be outputed at every step. 

* $\fbox{\text{Step 4:}}$ Steps 2 and 3 are repeated until every node is in a different community.

### 4.3 Community Coloring Implementation

```python
def get_random_color():
    # Generates a random color in hex format
    return "#{:06x}".format(random.randint(0, 0xFFFFFF))

# Function to draw the graph with communities
def draw_communities(G, communities):
    # Random colors for each community
    community_colors = [get_random_color() for _ in range(len(communities))]
    color_map = {}
    # Color to each node based on its community
    for idx, community in enumerate(communities):
        for node in community:
            color_map[node] = community_colors[idx]
    # Gets the color for each node in the graph
    node_colors = [color_map[node] for node in G.nodes()]

    plt.figure(figsize=(10, 8))
    pos = nx.spring_layout(G, seed=42)
    nx.draw(G, pos, with_labels=True, node_color=node_colors, node_size=500, font_size=8, edge_color='gray')
    plt.show()
```

### 4.4 Algorithm Implementation: Girvan-Newman Algorithm

```python
community_graph = G.copy()
max_modularity = 0
edges_max_modularity = []
n_communities_max_modularity = 0
numb_of_edges = len(G.edges())
nodes = list(community_graph.nodes())

for _ in range(numb_of_edges):
    # Betweenness centrality
    betweenness_centrality = nx.edge_betweenness_centrality(community_graph)
    edge_with_max_betweenness = max(betweenness_centrality.items(), key=lambda item: item[1])
    # Remove edge with maximum betweenness
    community_graph.remove_edge(edge_with_max_betweenness[0][0], edge_with_max_betweenness[0][1])
    components = list(nx.connected_components(community_graph))
    # Modularity calculation
    mod_sum = 0
    for component in components:
        for i in component:
            for j in component:
                if i != j:
                    a_i_j = 1 if community_graph.has_edge(i, j) else 0
                    mod_sum += a_i_j - (community_graph.degree(i) * community_graph.degree(j)) / (2 * numb_of_edges)
    
    modularity = mod_sum / (2 * numb_of_edges)
    # Check if the current modularity is the highest
    if modularity > max_modularity:
        max_modularity = modularity
        edges_max_modularity = edge_with_max_betweenness[0]
        n_communities_max_modularity = components

print(f"Max Modularity: {max_modularity}")
print(f"Number of Communities: {len(n_communities_max_modularity)}")
print(f"Communities: {n_communities_max_modularity}")
```

### 4.5 Community Drawing

```python
draw_communities(nx.Graph(G), n_communities_max_modularity)
```
