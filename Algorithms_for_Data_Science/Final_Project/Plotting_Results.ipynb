```python
import matplotlib.pyplot as plt

def plot_execution_time():
    # Define dataset information
    datasets = [
        {'name': 'mini_dataset1.csv', 'rows': 17, 'execution_time': '0.002s', 'color': 'blue'},
        {'name': 'mini_dataset2.csv', 'rows': 25, 'execution_time': '0.0032s', 'color': 'red'},
        {'name': '1000r', 'rows': 1000, 'execution_time': '44.7648s', 'color': 'green'},
        {'name': '10000r', 'rows': 10000, 'execution_time': '39min 33s', 'color': 'green'},
        {'name': '100000r', 'rows': 100000, 'execution_time': '2h 35min', 'color': 'green'},
        {'name': '500000r', 'rows': 500000, 'execution_time': '4h 55min', 'color': 'green'},
        {'name': 'Complete', 'rows': 1000000, 'execution_time': '7h 47min', 'color': 'green'},
    ]
    
    # Function to convert time strings to seconds
    def time_to_seconds(time_str):
        time_str = time_str.lower()
        seconds = 0
        if 'h' in time_str:
            parts = time_str.split('h')
            hours = float(parts[0].strip())
            seconds += hours * 3600
            time_str = parts[1] if len(parts) > 1 else ''
        if 'min' in time_str:
            parts = time_str.split('min')
            minutes = float(parts[0].strip())
            seconds += minutes * 60
            time_str = parts[1] if len(parts) > 1 else ''
        if 's' in time_str:
            parts = time_str.split('s')
            secs = float(parts[0].strip())
            seconds += secs
        return seconds
    
    # Extract rows and execution times in seconds
    rows = [ds['rows'] for ds in datasets]
    exec_times = [time_to_seconds(ds['execution_time']) for ds in datasets]
    colors = [ds['color'] for ds in datasets]
    names = [ds['name'] for ds in datasets]
    
    # Plotting
    plt.figure(figsize=(10, 6))
    
    # Plot all datasets
    for ds in datasets:
        exec_sec = time_to_seconds(ds['execution_time'])
        if ds['name'] == 'mini_dataset1.csv':
            plt.scatter(ds['rows'], exec_sec, color='blue', label='mini_dataset1.csv', s=100, marker='o')
            plt.annotate(ds['name'], (ds['rows'], exec_sec), textcoords="offset points", xytext=(0,10), ha='center', color='blue')
        elif ds['name'] == 'mini_dataset2.csv':
            plt.scatter(ds['rows'], exec_sec, color='red', label='mini_dataset2.csv', s=100, marker='s')
            plt.annotate(ds['name'], (ds['rows'], exec_sec), textcoords="offset points", xytext=(0,10), ha='center', color='red')
        else:
            plt.scatter(ds['rows'], exec_sec, color='green', label=ds['name'], s=50, marker='^')
            # Optionally annotate others, but may clutter the plot
    
    # Plot lines connecting all points
    sorted_datasets = sorted(datasets, key=lambda x: x['rows'])
    sorted_rows = [ds['rows'] for ds in sorted_datasets]
    sorted_exec_times = [time_to_seconds(ds['execution_time']) for ds in sorted_datasets]
    plt.plot(sorted_rows, sorted_exec_times, color='grey', linestyle='--', linewidth=1)
    
    # Create a custom legend to avoid duplicate labels
    handles = []
    labels = []
    for ds in datasets:
        if ds['name'] == 'mini_dataset1.csv':
            handles.append(plt.Line2D([0], [0], marker='o', color='w', label='mini_dataset1.csv',
                                      markerfacecolor='blue', markersize=10))
            labels.append('mini_dataset1.csv')
        elif ds['name'] == 'mini_dataset2.csv':
            handles.append(plt.Line2D([0], [0], marker='s', color='w', label='mini_dataset2.csv',
                                      markerfacecolor='red', markersize=10))
            labels.append('mini_dataset2.csv')
        elif ds['name'] in ['Complete', '500000r', '100000r', '10000r', '1000r']:
            handles.append(plt.Line2D([0], [0], marker='^', color='w', label=ds['name'],
                                      markerfacecolor='green', markersize=8))
            labels.append(ds['name'])
    plt.legend(handles, labels, loc='upper left', bbox_to_anchor=(1,1))
    
    # Set log scale for x-axis
    plt.xscale('log')
    
    plt.xlabel('Number of Rows')
    plt.ylabel('Execution Time (seconds)')
    plt.title('Execution Time vs. Number of Rows')
    plt.grid(True, which="both", ls="--", linewidth=0.5)
    plt.tight_layout()
    plt.show()

# Call the function to display the plot
plot_execution_time()
```

```python
import matplotlib.pyplot as plt

def plot_time_vs_memory():
    # Define dataset information
    datasets = [
        {'name': 'mini_dataset1.csv', 'execution_time': '0.002s', 'memory_usage': '0,007144', 'color': 'blue', 'marker': 'o'},
        {'name': 'mini_dataset2.csv', 'execution_time': '0.0032s', 'memory_usage': '0,007944', 'color': 'red', 'marker': 's'},
        {'name': '1000r', 'execution_time': '44.7648s', 'memory_usage': '63', 'color': 'green', 'marker': '^'},
        {'name': '10000r', 'execution_time': '39min 33s', 'memory_usage': '833', 'color': 'green', 'marker': '^'},
        {'name': '100000r', 'execution_time': '2h 35min', 'memory_usage': '2843', 'color': 'green', 'marker': '^'},
        {'name': '500000r', 'execution_time': '4h 55min', 'memory_usage': '3466', 'color': 'green', 'marker': '^'},
        {'name': 'Complete', 'execution_time': '7h 47min', 'memory_usage': '3616', 'color': 'green', 'marker': '^'},
    ]
    
    # Function to convert execution time strings to seconds
    def parse_time(time_str):
        time_str = time_str.lower()
        seconds = 0
        if 'h' in time_str:
            parts = time_str.split('h')
            hours = float(parts[0].strip())
            seconds += hours * 3600
            time_str = parts[1] if len(parts) > 1 else ''
        if 'min' in time_str:
            parts = time_str.split('min')
            minutes = float(parts[0].strip())
            seconds += minutes * 60
            time_str = parts[1] if len(parts) > 1 else ''
        if 's' in time_str:
            parts = time_str.split('s')
            secs = float(parts[0].strip())
            seconds += secs
        return seconds
    
    # Function to convert memory usage strings to float (handle comma as decimal separator)
    def parse_memory(mem_str):
        return float(mem_str.replace(',', '.'))
    
    # Extract execution times and memory usages
    exec_times = [parse_time(ds['execution_time']) for ds in datasets]
    memory_usages = [parse_memory(ds['memory_usage']) for ds in datasets]
    
    # Plotting
    plt.figure(figsize=(10, 6))
    
    for ds in datasets:
        exec_sec = parse_time(ds['execution_time'])
        mem_mb = parse_memory(ds['memory_usage'])
        if ds['name'] == 'mini_dataset1.csv':
            plt.scatter(mem_mb, exec_sec, color='blue', marker='o', s=100, label='mini_dataset1.csv')
            plt.annotate(ds['name'], (mem_mb, exec_sec), textcoords="offset points", xytext=(0,10), ha='center', color='blue', fontsize=9)
        elif ds['name'] == 'mini_dataset2.csv':
            plt.scatter(mem_mb, exec_sec, color='red', marker='s', s=100, label='mini_dataset2.csv')
            plt.annotate(ds['name'], (mem_mb, exec_sec), textcoords="offset points", xytext=(0,10), ha='center', color='red', fontsize=9)
        else:
            plt.scatter(mem_mb, exec_sec, color='green', marker='^', s=50, label=ds['name'] if ds['name'] not in ['mini_dataset1.csv', 'mini_dataset2.csv'] else "")
    
    # Connect larger datasets with a line
    large_datasets = [ds for ds in datasets if ds['name'] not in ['mini_dataset1.csv', 'mini_dataset2.csv']]
    sorted_large = sorted(large_datasets, key=lambda x: parse_memory(x['memory_usage']))
    sorted_mem = [parse_memory(ds['memory_usage']) for ds in sorted_large]
    sorted_exec = [parse_time(ds['execution_time']) for ds in sorted_large]
    plt.plot(sorted_mem, sorted_exec, color='grey', linestyle='--', linewidth=1)
    
    # Create a custom legend to avoid duplicate labels
    handles = [
        plt.Line2D([0], [0], marker='o', color='w', label='mini_dataset1.csv', markerfacecolor='blue', markersize=10),
        plt.Line2D([0], [0], marker='s', color='w', label='mini_dataset2.csv', markerfacecolor='red', markersize=10),
        plt.Line2D([0], [0], marker='^', color='w', label='Other Datasets', markerfacecolor='green', markersize=8)
    ]
    plt.legend(handles, [h.get_label() for h in handles], loc='upper left', bbox_to_anchor=(1,1))
    
    plt.xlabel('Memory Usage (MB)')
    plt.ylabel('Execution Time (seconds)')
    plt.title('Execution Time vs. Memory Usage')
    plt.grid(True, which="both", ls="--", linewidth=0.5)
    plt.tight_layout()
    plt.show()

# Call the function to display the plot
plot_time_vs_memory()
```
